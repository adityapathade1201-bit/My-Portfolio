```python
# 1. How many total matches are recorded in the dataset?
import numpy as np
total_matches = df['match_id'].nunique()
print(f"Total Matches: {total_matches}")
```

```python
# 2. List all the unique teams that have participated in the IPL according to this data.
unique_teams = np.unique(df[['batting_team', 'bowling_team']].values.flatten())
print("Unique Teams:", unique_teams)
```

```python
# 3. What is the maximum number of runs scored by a batter in a single delivery in this dataset?
max_runs = df['batter_runs'].max()
print(f"Maximum runs in a single ball: {max_runs}")
```

```python
# 4. Find the total number of runs scored in the entire history of the dataset.
total_runs = df['batter_runs'].sum()
print(f"Total Runs Scored: {total_runs}")
```

```python
# 5. Which player has faced the most number of balls?
balls_faced = df['new_batter'].value_counts().idxmax()
print(f"Player who faced most balls: {balls_faced}")
```

```python
# 6. Calculate the average runs scored per match.
avg_runs_per_match = df.groupby('match_id')['batter_runs'].sum().mean()
print(f"Average runs per match: {avg_runs_per_match:.2f}")
```

```python
# 7. How many wickets were taken in total? (Assuming bowler_wicket column tracks this)
total_wickets = df['bowler_wicket'].sum()
print(f"Total Wickets: {total_wickets}")
```

```python
# 8. Find the match dates for all matches played by 'Kolkata Knight Riders'.
kkr_dates = df[df['batting_team'] == 'Kolkata Knight Riders']['date'].unique()
print("KKR Match Dates:", kkr_dates[:10]) # Displaying first 10
```

```python
# 9. Identify the top 5 batters with the highest total runs.
top_5_batters = df.groupby('new_batter')['batter_runs'].sum().sort_values(ascending=False).head(5)
print("Top 5 Batters:")
print(top_5_batters)
```

```python
# 10. What is the total number of 'sixes' (6 runs in a single ball) scored?
total_sixes = len(df[df['batter_runs'] == 6])
print(f"Total Sixes: {total_sixes}")
```

```python
# 11. Count the number of matches played each year.
# Note: Assuming 'date' was converted to datetime previously
df['date'] = pd.to_datetime(df['date'])
matches_per_year = df.groupby(df['date'].dt.year)['match_id'].nunique()
print("Matches per Year:")
print(matches_per_year)
```

```python
# 12. Which batting team has scored the highest number of runs in a single match innings?
highest_innings_score = df.groupby(['match_id', 'innings', 'batting_team'])['batter_runs'].sum().reset_index().max()
print("Highest Innings Score Details:")
print(highest_innings_score)
```

```python
# 13. Find the percentage of balls that resulted in a wicket.
wicket_percentage = (df['bowler_wicket'].sum() / len(df)) * 100
print(f"Wicket Percentage: {wicket_percentage:.2f}%")
```

```python
# 14. List the matches played in the first innings where the score exceeded 200 runs.
high_scores = df[df['innings'] == 1].groupby('match_id')['batter_runs'].sum()
matches_above_200 = high_scores[high_scores > 200]
print(f"Number of 200+ scores in 1st innings: {len(matches_above_200)}")
```

```python
# 15. Determine the average number of runs scored in the first over (over 0) of all matches.
first_over_avg = df[df['over'] == 0].groupby('match_id')['batter_runs'].sum().mean()
print(f"Average runs in the first over: {first_over_avg:.2f}")
```

```python
# 16. Find the player names who were out (striker_out is True).
players_out = df[df['striker_out'] == True]['new_batter'].unique()
print(f"Number of unique players who got out: {len(players_out)}")
```

```python
# 17. Compare total runs scored by 'Royal Challengers Bangalore' vs 'Chennai Super Kings'.
rcb_runs = df[df['batting_team'] == 'Royal Challengers Bangalore']['batter_runs'].sum()
csk_runs = df[df['batting_team'] == 'Chennai Super Kings']['batter_runs'].sum()
print(f"RCB Total Runs: {rcb_runs}")
print(f"CSK Total Runs: {csk_runs}")
```

```python
# 18. Filter the dataset to show only matches played after 2020.
recent_matches = df[df['date'].dt.year > 2020]
print(f"Number of ball-by-ball records after 2020: {len(recent_matches)}")
```

```python
# 19. Identify the top 5 bowlers who have bowled the most dot balls (0 runs).
# We define a dot ball as a ball where no runs (batter or extras) were scored.
if 'extras_runs' in df.columns:
    dot_balls_df = df[(df['batter_runs'] == 0) & (df['extras_runs'] == 0)]
else:
    # If a specific extras column isn't found, we use batter_runs as the primary metric
    dot_balls_df = df[df['batter_runs'] == 0]

dot_balls = dot_balls_df.groupby('bowler').size().sort_values(ascending=False).head(5)
print("Top 5 Bowlers with most dot balls:")
print(dot_balls)
```

```python
# 20. Create a pivot table showing total runs by batting team for each inning.
pivot_runs = df.pivot_table(index='batting_team', columns='innings', values='batter_runs', aggfunc='sum')
display(pivot_runs)
```
